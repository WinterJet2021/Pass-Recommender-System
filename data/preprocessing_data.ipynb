#### Import Libraries

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
```

```python
file_path = "user_interests_50_dataset.csv"  
df = pd.read_csv(file_path)
```

```python
location_mapping = pd.read_csv("location_mapping.csv")
nationality_mapping = pd.read_csv("nationality_mapping.csv")
language_mapping = pd.read_csv("language_mapping.csv")
```

```python
location_dict = dict(zip(location_mapping["Location"], location_mapping["ID"]))
nationality_dict = dict(zip(nationality_mapping["Nationality"], nationality_mapping["ID"]))
language_dict = dict(zip(language_mapping["Language"], language_mapping["ID"]))
```

```python
activity_columns = ["Basketball", "Yoga", "Hiking", "Cycling", "Gym", "Swimming", "Dancing", "Running", "Music", "Photography"]
```

```python
df[activity_columns]
```

```python
min_value, max_value = 1, 5
df[activity_columns] = (df[activity_columns] - min_value) / (max_value - min_value)
df[activity_columns] = df[activity_columns].round(2)
```

```python
df[activity_columns]
```

```python
user_score_min, user_score_max = 1, 10
df["User Score"] = (df["User Score"] - user_score_min) / (user_score_max - user_score_min)
df["User Score"] = df["User Score"].round(2)
```

```python
days_of_week = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
df["Availability_Vector"] = df["Availability"].apply(lambda x: "[" + ", ".join(str(1 if day in x.split(", ") else 0) for day in days_of_week) + "]")
df.drop(columns=["Availability"], inplace=True)
```

```python
gender_mapping = {"Male": 0, "Female": 1, "Other": 2, "Prefer not to say": 3}
df["Gender"] = df["Gender"].map(gender_mapping)
```

```python
df["Location_ID"] = df["Location"].map(location_dict)
df["Nationality_ID"] = df["Nationality"].map(nationality_dict)
```

```python
def map_languages(languages):
    if pd.isna(languages):
        return ""
    return [language_dict.get(lang.strip(), None) for lang in languages.split(",") if lang.strip() in language_dict]

df["Language_IDs"] = df["Languages Spoken"].apply(map_languages)
```

```python
df.drop(columns=["Location", "Nationality", "Languages Spoken"], inplace=True)
```

```python
df.to_csv("../Dataset/Pre_data/preprocessed_50user_data.csv", index=False)
```
